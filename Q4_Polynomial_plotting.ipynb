```python
import numpy as np
import matplotlib.pyplot as plt

# Set the random seed
np.random.seed(12)

# Generate random coefficients for a 5th degree polynomial
coefficients = np.random.randint(-10, 10, 6)  # Random integers between -10 and 10 inclusive

# Define the polynomial function
def polynomial(x, coeffs):
    return np.polyval(coeffs, x)

# Generate x values
x_values = np.linspace(-5, 5, 1000)

# Generate y values using the polynomial function
y_values = polynomial(x_values, coefficients)

# Plot the polynomial
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Polynomial')

# Find maxima and minima
maxima_indices = np.where(np.diff(np.sign(np.diff(y_values))) < 0)[0] + 1
minima_indices = np.where(np.diff(np.sign(np.diff(y_values))) > 0)[0] + 1

maxima_x = x_values[maxima_indices]
maxima_y = y_values[maxima_indices]
minima_x = x_values[minima_indices]
minima_y = y_values[minima_indices]

# Annotate maxima and minima
for i in range(len(maxima_x)):
    plt.annotate(f'Maxima {i+1}', xy=(maxima_x[i], maxima_y[i]), xytext=(maxima_x[i]+0.5, maxima_y[i]+20),
                 arrowprops=dict(facecolor='black', arrowstyle='->'))

for i in range(len(minima_x)):
    plt.annotate(f'Minima {i+1}', xy=(minima_x[i], minima_y[i]), xytext=(minima_x[i]+0.5, minima_y[i]-20),
                 arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.title('Plot of 5th Degree Polynomial with Maxima and Minima')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
